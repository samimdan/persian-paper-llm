```python
from docx import Document

def read_word(path):
    doc = Document(path)
    paragraphs = []
    for p in doc.paragraphs:
        text = p.text.strip()
        if len(text) > 0:
            paragraphs.append(text)
    return paragraphs

paragraphs = read_word("thesis1.docx")
print("تعداد پاراگراف‌ها:", len(paragraphs))
print(paragraphs[0])
```

```python
def chunk_text(paragraphs, max_words=200):
    chunks = []
    current = []

    for p in paragraphs:
        words = p.split()
        if len(current) + len(words) <= max_words:
            current.extend(words)
        else:
            chunks.append(" ".join(current))
            current = words

    if current:
        chunks.append(" ".join(current))

    return chunks

chunks = chunk_text(paragraphs)
print("تعداد چانک‌ها:", len(chunks))
print(chunks[0])
```
